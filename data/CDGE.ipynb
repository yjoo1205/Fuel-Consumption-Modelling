### DO-2019

```python
from io import StringIO
import pandas as pd
from azure.storage.blob import BlobClient, BlobServiceClient
import numpy as np
import pandas as pd
import datetime
```

```python
#!pip install azure-storage-blob --pre
STORAGEACCOUNTNAME= 'cdgtemp01'    
CONTAINERNAME= 'cdge'
BLOBNAME= 'DO-2019.csv' #replace this string with the data file name and extension
sastoken = '?sv=2019-12-12&si=interns&sr=c&sig=dv3uO8P8oMfyvFozONMJWwIDKzMQLoQR4RI3y8spQC0%3D'
url='https://cdgtemp01.blob.core.windows.net'

blob_service_client = BlobServiceClient(account_url=url,credential=sastoken)
blob_client = blob_service_client.get_blob_client(container=CONTAINERNAME, blob=BLOBNAME)
blob = blob_client.download_blob().content_as_text(encoding='iso-8859-1')

#store as a dataframe
df = pd.read_csv(StringIO(blob))
df.head(25)
```

```python
df.info()
```

```python
df.columns = ['Delivery ID', 'Ship-to Key', 'Material', 'Dlv.qty', 'Total Weight', 'WUn', 'Volume', 'Description', 
              'Deliv.date', 'TrpPlanDt', 'Pick Date', 'Net weight', 'Dlv.qty.1', 'Name of the ship-to party']
```

```python
df[df['Ship-to Key'].isnull()]
```

```python
df = df.dropna(subset=['Ship-to Key'])
```

```python
df.info()
```

```python
df = df.reset_index()
```

```python
# dulicated data -> at that one day has many trips to that one location or error?
# some  data total weight = 0 -> error?
df[df.duplicated(keep=False)]
```

```python
df['Ship-to Key'].apply(float.is_integer).all()
```

```python
df['Ship-to Key'] = df['Ship-to Key'].astype('int64')
```

```python
def convert_to_float(value):
    try:
        value = float(value)
    except ValueError:
        value = value.replace(',','')
        value = float(value)
    return value
```

```python
df['Dlv.qty'] = df['Dlv.qty'].apply(convert_to_float)
```

```python
df['Total Weight'] = df['Total Weight'].apply(convert_to_float)
```

```python
df['Volume'] = df['Volume'].apply(convert_to_float)
```

```python
df['Net weight'] = df['Net weight'].apply(convert_to_float)
```

```python
df = df.drop('index', axis=1)
```

```python
df.info()
```

```python
len(df[df['WUn'] == 'KG'])
```

```python
df[df['Deliv.date'].isnull()]
```

```python
df = df.dropna(subset=['Deliv.date'])
```

```python
df['Deliv.date'] = df['Deliv.date'].apply(lambda x: datetime.datetime.strptime(str(x), '%d.%m.%Y').date())
df['TrpPlanDt'] = df['TrpPlanDt'].apply(lambda x: datetime.datetime.strptime(str(x), '%d.%m.%Y').date())
df['Pick Date'] = df['Pick Date'].apply(lambda x: datetime.datetime.strptime(str(x), '%d.%m.%Y').date())
```

```python
df['Deliv.date'] = df['Deliv.date'].astype('datetime64[ns]')
df['TrpPlanDt'] = df['TrpPlanDt'].astype('datetime64[ns]')
df['Pick Date'] = df['Pick Date'].astype('datetime64[ns]')
```

```python
df['Delivery ID'] = df['Delivery ID'].astype(int)
```

```python
df.head()
```

```python
df.info()
```
